# **Bank Marketing Campaign**

## **A.Business Problem Understanding**

### Business Problem
Bank menghadapi tantangan dalam efektivitas kampanye pemasaran untuk produk deposito berjangka (term deposit). Berdasarkan data historis, kampanye yang dilakukan secara massal kepada semua nasabah membutuhkan biaya operasional yang tinggi (biaya telepon/cellular contact, tenaga marketing, waktu) namun tingkat konversi nasabah yang berhasil membuka deposito (deposit: yes) masih rendah.

Observasi dari Data:

- Terdapat variasi karakteristik nasabah yang sangat beragam (umur 31-62 tahun, berbagai pekerjaan dari admin hingga management)
- Balance nasabah sangat bervariasi (dari negatif -87 hingga 13,901)
- Beberapa nasabah memiliki housing loan dan personal loan yang berbeda-beda
- Campaign dilakukan dengan frekuensi berbeda (1-4 kali per nasabah)
- Hasil kampanye sebelumnya (poutcome) mayoritas "unknown" dengan beberapa "success"
- Kontak dilakukan di bulan yang berbeda-beda (jan-nov)
- pdays menunjukkan -1 (never contacted) atau angka tertentu (184, 352)

Dampak Bisnis:

Biaya kampanye pemasaran tinggi dengan ROI rendah
Tingkat konversi pelanggan yang membuka deposito rendah (banyak deposit: no)
Pengalaman pelanggan buruk karena dihubungi berulang kali padahal tidak berminat
Kehilangan peluang untuk fokus pada nasabah potensial yang benar-benar berminat

### **Stakeholder**
Pemangku kepentingan utama dalam analisis ini adalah tim pemasaran dan penjualan bank, yang bertanggung jawab langsung dalam pelaksanaan dan optimalisasi kampanye deposito berjangka.
Dengan memanfaatkan wawasan dari analisis ini, tim tersebut dapat mengidentifikasi dan memprioritaskan nasabah yang paling berpotensi untuk berlangganan deposito, sehingga strategi pemasaran dapat dilakukan secara lebih terarah dan efisien.

Pendekatan ini tidak hanya mengurangi biaya akibat menghubungi nasabah yang tidak tertarik, tetapi juga meningkatkan tingkat konversi kampanye, memastikan bahwa sumber daya digunakan secara optimal untuk mencapai hasil maksimal.

### Goals
**Primary Goal:**
Meningkatkan efektivitas kampanye pemasaran deposito berjangka dengan meningkatkan conversion rate dari kondisi saat ini menjadi 25-30% dalam 6 bulan ke depan melalui targeting yang lebih akurat.

**Secondary Goals:**
- Mengurangi biaya kampanye pemasaran hingga 30-40% dengan menargetkan nasabah yang lebih tepat
- Mengoptimalkan jumlah campaign contact per nasabah (saat ini 1-4 kali)
- Meningkatkan efisiensi tim marketing dengan mengurangi kontak yang tidak produktif
- Meningkatkan kepuasan nasabah dengan mengurangi kontak yang tidak relevan
- Mengidentifikasi segmen nasabah dengan potensi tertinggi (berdasarkan job, age, balance, dll)
- Menentukan timing optimal untuk kampanye (bulan terbaik)
- Meningkatkan total nilai deposito yang berhasil dihimpun sebesar 40-50%

### **Metric Evaluation**

Rata budget marketing yang di terapkan bank (Sumber : [ABA Banking Journal](https://bankingjournal.aba.com/2025/01/bank-marketing-budget-and-staffing-considerations-for-2025/))



#### Marketing Budget
- Diperkiraan bahwa total aset dari bank yaitu $300MM-499MM
- Sehingga estimasi budget marketingnya yaitu $205K- $ 200k
- Dimana diasumsikan bahwa target untuk new Customer adalah 34%
- Sehingga total budget untuk dataset ini diperkirakan sekitar $200K X 34% = $68000
- Dari Budget Tersebut diperkirakan untuk tipe marketing ini yaitu 50% dari budget atau $34000
- Target customer baru untuk deposito adalah 100000 sehingga budget marketing per nasabah yaitu $3.4/ per nasabah

#### Cost Profit Evaluation
Jika dalam data rata-rata saldo bank yaitu $550 dan diasumsikan dimasukan deposito sekitar 40% atau sekitar $220
- Bunga Deposito perbulan yaitu 3% maka
-- $220 X 3%=  $6.6 ( merupakan biaya  yang harus dikembalikan ke customer)
- Dan dari deposito tersebut digunakan bank untuk investasi dll dan mendapat keuntungan 10%
-- $220 X 10 %= $22

sehingga perhiungan keunutukan dari setiap matriks seperti berikut
- TP : $ 22 - $ 6.6 - $ 3.4 = + $ 12 (net benefit)
- TN : + $ 3.4 (hemat biaya operasional)
- FP : - $ 3.4 (rugi biaya operasional)
- FN : - $ 12 (kehilangan peluang net benefit)



| **Prediksi vs Aktual**   | **Definisi**                                                                           | **Dampak**                                                                                 | **Perhitungan Manfaat Bersih (Net Benefit)**             |
| ------------------------ | -------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------ | -------------------------------------------------------- |
| **True Positives (TP)**  | Nasabah yang **benar diprediksi akan berlangganan deposito**.                          | 🟢 **Keberhasilan pemasaran langsung** – Memaksimalkan akuisisi deposito dan pendapatan.   | ✅ **Keuntungan:** TP × *Profit per TP* = TP × **$8.15** |
| **False Positives (FP)** | Nasabah yang **diprediksi akan berlangganan**, tetapi **sebenarnya tidak**.            | ❌ **Biaya pemasaran yang tidak perlu** – Mengakibatkan upaya promosi yang sia-sia.         | 🔻 **Kerugian:** FP × *Cost per FP* = FP × **$3.4**      |
| **True Negatives (TN)**  | Nasabah yang **benar diprediksi tidak berlangganan deposito**.                         | ✅ **Penghematan biaya pemasaran** – Menghindari kontak dengan nasabah yang tidak tertarik. | ⚖️ **Netral:** TN × *Profit per TN* = TN × **$3.4**        |
| **False Negatives (FN)** | Nasabah yang **salah diprediksi tidak berlangganan**, padahal sebenarnya berlangganan. | 🔴 **Peluang hilang** – Potensi kehilangan nasabah yang seharusnya bisa berlangganan.      | 🔻 **Kerugian:** FN × *Loss per FN* = FN × **$8.15**    |

# Preprocess

```python
preprocess_transformer = ColumnTransformer([
    ('binary', BinaryEncoder(), ['job']),
    ('onehot', OneHotEncoder(drop='first', sparse_output= False), ['housing', 'loan', 'contact']),
    ('robust', RobustScaler(), numcols)
], remainder= 'passthrough').set_output(transform='pandas')

preprocess_transformer
```

```python
preprocess_transformer.fit(Xtrain,ytrain)
```

# **Model Selection**

```python
# Define the models

# Logistic Regression
logreg = LogisticRegression(random_state =0)

# KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Decision Tree
dtree = DecisionTreeClassifier( random_state=0)

# RandomForest
rf = RandomForestClassifier(random_state=0)

# Adaboost
adaboost = AdaBoostClassifier(random_state=0)

# Gradientboost
gradboost = GradientBoostingClassifier(random_state=0)

# ExtremeGradientBoost
xgboost = XGBClassifier(random_state=0)

```

```python
models = [logreg, knn, dtree, rf, adaboost, gradboost, xgboost]

list_meanf1 = []
list_stdf1 = []
list_all_scoref1 = []


for model in models :

    pipe_model = Pipeline([
        ('preprocess', preprocess_transformer),
        ('model', model)
    ])

   
    skfold = StratifiedKFold(n_splits = 5)
    
    model_cv = cross_val_score(
        estimator=pipe_model,
        X = Xtrain,    
        y = ytrain,
        cv = skfold,
        scoring = 'f1'
)
    
    list_meanf1.append(model_cv.mean())
    list_stdf1.append(model_cv.std())
    list_all_scoref1.append(model_cv.round(4))
```

```python
df_cv = pd.DataFrame()
df_cv["model"] = models
df_cv["All Score"] = list_all_scoref1
df_cv["F1 Mean"] = list_meanf1
df_cv["F1 Std"] = list_stdf1

df_cv.sort_values(["F1 Mean", "F1 Std"], 
                  ascending=[False, True])
```

# **Hyperparameter Tuning**

```python
models = {
    'Gradient Boosting': {
        'model': GradientBoostingClassifier(random_state=0),
        'param_grid': {
            'model__n_estimators': range(50, 100),
            'model__max_depth': range(3, 12),
            'model__learning_rate': np.logspace(np.log10(0.01), np.log10(0.20), num=50),
            'model__min_samples_split': range(20, 80, 5),  
            'model__min_samples_leaf': range(10, 40, 3),
            'model__subsample': np.linspace(0.6, 1.0, 5), 
            'model__max_features': ['sqrt', 'log2', None],  
            'model__loss': ['log_loss', 'exponential'],  
            'model__ccp_alpha': np.linspace(0.0, 0.01, 5), 
}},
'XGBoost': {
        'model': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=0),
        'param_grid': {
            'model__n_estimators': range(50, 100),
            'model__max_depth': range(3, 12),  
            'model__learning_rate': np.logspace(np.log10(0.01), np.log10(0.20), num=50),       
            'model__scale_pos_weight': range(10, 40),  # For imbalanced data                  
            'model__min_child_weight': range(10, 40),             
            'model__max_delta_step': (1, 3, 5),    
}}}
```

```python
results = []

stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

for model_name, config in models.items():
    print(f"\nTuning hyperparameters for: {model_name}")

    pipeline = Pipeline(steps=[
        ('preprocessing', preprocess_transformer),
        ('model', config['model'])
    ])

    random_search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=config['param_grid'],
        n_iter=50,  
        cv=stratified_kfold,  
        scoring='f1',
        verbose=1,
        n_jobs=-1,
        random_state=0
    )
   
    random_search.fit(Xtrain, ytrain)
   
    results.append({
        'model': model_name,
        'best_params': random_search.best_params_,
        'best_f1_score': random_search.best_score_
    })

df_hypertuned = pd.DataFrame(results)
```

## Threshold Optimization

```python
y_proba = bestmodel.predict_proba(Xtest)[:, 1]
thresholds = np.arange(0.1, 1.0, 0.001)

results = []

for threshold in thresholds:
    y_pred_adj = (y_proba >= threshold).astype(int)

    recall = recall_score(ytest, y_pred_adj)
    precision = precision_score(ytest, y_pred_adj)
    f1 = f1_score(ytest, y_pred_adj)
    f2 = fbeta_score(ytest, y_pred_adj, beta=2)
    accuracy = accuracy_score(ytest, y_pred_adj)

    results.append((threshold, recall, precision, f1, f2, accuracy))

df_thresholds = pd.DataFrame(results, columns=['Threshold', 'Recall', 'Precision', 'F1-Score', 'F2-Score', 'Accuracy'])

best_thresholds = df_thresholds.sort_values(by='F1-Score', ascending=False).head(10)
print(best_thresholds)

best_threshold = best_thresholds.iloc[0, 0]
print(f"\nBest Threshold berdasarkan F1-Score: {best_threshold:.3f}")
```

## Final Model Evaluation

```python
def cost(cm):
    tn, fp, fn, tp = cm.reshape(-1)
    total_sampel = tn + fp + fn + tp
    
    tp_cost = tp * 12
    tn_cost = tn * 3.4
    fp_cost = fp * 3.4
    fn_cost= fn * 12
    
    # Hasil bersih
    total_keuntungan = tp_cost + tn_cost
    total_biaya = fp_cost + fn_cost
    laba_bersih = total_keuntungan - total_biaya
    
    summary = f'''
    Total Sampel: {total_sampel}
    
    Keuntungan:
    - Pendapatan TP: $ {tp_cost:,.2f} (dari {tp})
    - Penghematan TN: $ {tn_cost:,.2f} (dari {tn})
    - Total Keuntungan: $ {total_keuntungan:,.2f}
    
    Biaya:
    - Biaya FP: $ {fp_cost:,.2f} (dari {fp})
    - Biaya FN: $ {fn_cost:,.2f} (dari {fn})
    - Total Biaya: $ {total_biaya:,.2f}
    
    Laba Bersih: $ {laba_bersih:,.2f}
    '''
    
    return summary
```
